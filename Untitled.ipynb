```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import PCA
import joblib  # Untuk menyimpan model
import requests
import re
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
```

```python
ENGLISH_STOP_WORDS = set(stopwords.words('english'))
INDONESIAN_STOP_WORDS = set(stopwords.words('indonesian'))
ALL_STOP_WORDS = ENGLISH_STOP_WORDS.union(INDONESIAN_STOP_WORDS)
```

```python
# --- 1. Memuat Data ---
data = pd.read_csv("data/updatemergedata.csv")

# --- 2. Preprocessing Data ---
# Encode categorical variables
label_encoder = LabelEncoder()
# Definisikan urutan level yang diinginkan
level_order = ['Dasar', 'Pemula', 'Menengah', 'Mahir', 'Profesional']
# Konversi kolom 'Level' menjadi kategori dengan urutan spesifik
data['Level'] = pd.Categorical(data['Level'], categories=level_order, ordered=True)
# Encode 'Level' menjadi nilai numerik berdasarkan urutan
data['Level'] = data['Level'].cat.codes
data['Price'] = data['Price'].map({'Berbayar': 1, 'Gratis': 0})  # Encode Price into binary (1: Berbayar, 0: Gratis)

# Gabungkan fitur teks menjadi satu kolom
indonesian_stopwords = stopwords.words('indonesian')
data['Combined Summary'] = (
    data['Learning Path'] + ' ' +
    data['Learning Path Summary'] + ' ' +
    data['Course Name_x'] + ' ' +
    data['Course Summary']
)

# Menghilangkan stopwords
data['Combined Summary'] = data['Combined Summary'].apply(lambda x: ' '.join([word for word in word_tokenize(x) if word not in indonesian_stopwords]))
```

```python
    # --- 3. Vektorisasi Fitur Teks ---
    # Menggunakan TfidfVectorizer untuk vektorisasi teks
    vectorizer = TfidfVectorizer(stop_words=indonesian_stopwords)
    X = vectorizer.fit_transform(data['Combined Summary'])
```

```python

# --- 4. Clustering dengan KMeans ---
# Menentukan jumlah cluster dengan metode Elbow
distortions = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(X)
    distortions.append(kmeans.inertia_)

# Plot Elbow Curve untuk menentukan k optimal
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), distortions, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters')
plt.ylabel('Distortion')
plt.grid(True)
plt.show()
```

```python
# Berdasarkan plot elbow, pilih jumlah cluster yang optimal
optimal_clusters = 9  #

# Melakukan clustering menggunakan KMeans
kmeans = KMeans(n_clusters=optimal_clusters, random_state=0)
data['Cluster'] = kmeans.fit_predict(X)

# --- 5. Menyimpan Model ---
# Menyimpan model KMeans dan vectorizer yang telah dilatih
joblib.dump(kmeans, 'kmeans_model.pkl')
joblib.dump(vectorizer, 'vectorizer.pkl')
```

```python
def recommend_courses(user_skill, user_language, github_username, similarity_threshold=0.3, user_weight=0.7, github_weight=0.3, min_recommendations=6, max_recommendations=10):
    # Fetch GitHub data and extract keywords
    github_keywords = fetch_github_keywords(github_username)
    
    # Gabungkan input pengguna dengan data GitHub
    if github_keywords:
        print("\nGitHub data included in recommendations.")
        combined_input = f"{user_skill} {user_language} {github_keywords}"
    else:
        print("\nNo valid GitHub data found. Proceeding with user input only.")
        combined_input = f"{user_skill} {user_language}"
    
    # Vektorisasi input pengguna dan data GitHub
    user_vector = vectorizer.transform([f"{user_skill} {user_language}"])
    combined_vector = vectorizer.transform([combined_input])
    
    # Prediksi cluster untuk input pengguna dan data gabungan
    user_cluster = kmeans.predict(user_vector)[0]
    combined_cluster = kmeans.predict(combined_vector)[0]
    
    # Filter data berdasarkan cluster gabungan
    cluster_data = data[(data['Cluster'] == user_cluster) | (data['Cluster'] == combined_cluster)].copy()
    
    # Hitung kesamaan kosinus untuk input pengguna dan data gabungan
    cluster_data['User Similarity'] = cosine_similarity(user_vector, X[cluster_data.index]).flatten()
    cluster_data['Combined Similarity'] = cosine_similarity(combined_vector, X[cluster_data.index]).flatten()
    
    # Gabungkan skor kesamaan dengan pembobotan
    cluster_data['Final Similarity'] = (
        user_weight * cluster_data['User Similarity'] + 
        github_weight * cluster_data['Combined Similarity']
    )
    
    # Filter rekomendasi berdasarkan similarity threshold
    filtered_recommendations = cluster_data[cluster_data['Final Similarity'] > similarity_threshold]
    
    # Jika jumlah rekomendasi kurang dari minimum, gunakan semua data cluster
    if len(filtered_recommendations) < min_recommendations:
        filtered_recommendations = cluster_data.copy()
        print("\nThreshold ignored. Displaying all available recommendations in cluster.")
    
    # Urutkan rekomendasi berdasarkan Level (ascending) dan Final Similarity (descending)
    filtered_recommendations = filtered_recommendations.sort_values(by=['Level', 'Final Similarity'], ascending=[True, False])
    
    # Batasi jumlah rekomendasi (maksimal 10) dan hilangkan duplikat nama kursus
    recommended_courses = filtered_recommendations.drop_duplicates(subset=['Course Name_x']).head(max_recommendations)
    
    # Tampilkan hasil rekomendasi
    print("\nRecommended Learning Paths:")
    print(recommended_courses[['Course Name_x', 'Level', 'Final Similarity']])
```

```python
# --- 7. Visualisasi Hasil Clustering ---
# Menggunakan PCA untuk mereduksi dimensi menjadi 2 untuk visualisasi
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(X.toarray())

# Menampilkan hasil PCA dalam plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=reduced_data[:, 0], y=reduced_data[:, 1], hue=data['Cluster'], palette='Set2', s=100, alpha=0.7)
plt.title('Clustering Learning Paths')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
def fetch_github_keywords(github_username):
    # URL untuk GitHub API
    url = f"https://api.github.com/users/{github_username}/repos"
    try:
        response = requests.get(url)
        response.raise_for_status()  # Cek jika ada error dari request
        repos = response.json()

        if not repos:
            print(f"No repositories found for user: {github_username}")
            return ""

        # Ambil informasi yang relevan dan gabungkan
        keywords = []
        for repo in repos:
            text = f"{repo['name']} {repo['description'] or ''} {repo['language'] or ''}"
            # Hilangkan karakter non-alfabet dan ubah ke huruf kecil
            cleaned_text = re.sub(r"[^a-zA-Z\s]", "", text).lower()
            # Tokenisasi teks
            tokens = word_tokenize(cleaned_text)
            # Hilangkan stopwords
            filtered_tokens = [word for word in tokens if word not in ALL_STOP_WORDS]
            keywords.extend(filtered_tokens)
        
        # Hilangkan kata duplikat
        unique_keywords = list(set(keywords))
        return " ".join(unique_keywords)
    except requests.exceptions.RequestException as e:
        print(f"Failed to fetch data from GitHub API: {e}")
        return ""
```
