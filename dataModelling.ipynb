```python
dataupdate = pd.read_csv(r"data\updatemergedata.csv")
dataupdate.head()
```

```python
# Mengonversi kolom 'Rating' menjadi numerik, dan mengubah nilai yang tidak bisa dikonversi menjadi NaN
dataupdate['Rating'] = pd.to_numeric(dataupdate['Rating'], errors='coerce')

# Menghapus baris yang memiliki nilai NaN di kolom 'Rating' (jika diperlukan)
dataupdate = dataupdate.dropna(subset=['Rating'])


average_rating = dataupdate['Rating'].mean()
print (average_rating)
```

```python
dataupdate['Rating'].fillna(0, inplace=True)
```

```python
# Daftar Course ID yang ingin diganti
course_ids_to_update = [313, 32, 153, 237, 281, 287, 312, 317, 335, 445, 505, 545, 550, 570, 580, 605, 620, 653, 658, 663, 697, 698, 703]

# Mengganti nilai Learning Path dan Learning Path Summary untuk Course ID tertentu
dataupdate.loc[dataupdate['Course ID'].isin(course_ids_to_update), ['Learning Path', 'Learning Path Summary']] = 'non learning path'

# Menampilkan hasil untuk memastikan perubahan
print(dataupdate[dataupdate['Course ID'].isin(course_ids_to_update)].head())
```

```python
import pandas as pd

# Membaca data dari file CSV
file_path = 'data/updatemergedata.csv'  # Ganti dengan lokasi file Anda
data = pd.read_csv(file_path)

# Menggabungkan judul modul/tutorial berdasarkan Course ID
grouped_data = data.groupby(['Course ID']).agg({
    'Learning Path ID': 'first',
    'Learning Path': 'first',
    'Learning Path Summary': 'first',
    'Course Name_x': 'first',
    'Course Summary': 'first',
    'Price': 'first',
    'Level': 'first',
    'Total Jam Belajar': 'first',
    'Rating': 'first',
    'Jumlah Enrollment': 'first',
    'Jumlah Modul/Tutorial': 'first',
    'Judul Modul/Tutorial': lambda x: ', '.join(x)
}).reset_index()

# Menyimpan hasil ke file baru
output_file = 'data/grouped_data.csv'  # Lokasi file output
grouped_data.to_csv(output_file, index=False)

print(f"Data berhasil diproses dan disimpan ke {output_file}")
```
